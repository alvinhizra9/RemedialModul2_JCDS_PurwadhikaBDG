If you realize, some columns that are numerical is listed as object data type. This is because in the data there is '?' value that become the placeholder of Null value. Now, replace all the '?' in every columns as NaN value (np.nan).

```python
for i in df.columns:
    df[i] = df[i].apply(lambda x : np.nan if x == '?' else x)
df.info()
```

```python
pd.DataFrame({
    'dataFeatures' : df.columns,
    'dataType' : [df[i].dtype for i in df.columns],
    'null' : [df[i].isna().sum() for i in df.columns],
    'nullPct' : [df[i].isna().sum()/len(df)*100 for i in df.columns],
    'unique' : [df[i].nunique() for i in df.columns],
    'uniqueSample' : [df[i].sample(2,random_state=101).values for i in df.columns]
})
```

There are 2 columns that would be nice if they are represented as numerical value rather than written; Num-of-Doors and Cylinders. For this 2 columns, replace the string as numerical value (Ex: if it is two then replace it with 2, etc.)

```python
def applynum(x):
    list_alpha_num = ['zero','one','two','three','four','five','six','seven','eight','nine','ten','eleven','twelve']
    if pd.isna(x):
        return x
    else:
        return list_alpha_num.index(x)
```

```python
df['Num-of-Doors'] = df['Num-of-Doors'].apply(applynum)
df['Cylinders'] = df['Cylinders'].apply(applynum)
```

Now convert these following columns into float object:
- Num-of-Doors
- Cylinders
- Bore
- Stroke
- Horsepower
- Peak-rpm

```python
df['Num-of-Doors'] = df['Num-of-Doors'].astype(float)
df['Cylinders'] = df['Cylinders'].astype(float)
df['Bore'] = df['Bore'].astype(float)
df['Stroke'] = df['Stroke'].astype(float)
df['Horsepower'] = df['Horsepower'].astype(float)
df['Peak-rpm'] = df['Peak-rpm'].astype(float)
```

```python
df.info()
```

We do not need the Symboling and Normalized Columns, so drop these 2 columns. Then, filter the data to acquire the only the row that contain NaN.

```python
df.drop(['Symboling','Normalized-Loss'],axis=1,inplace=True)
```

```python
list_index_na = []
for i in df.columns:
    for j,k in enumerate(df[i]):
        if pd.isna(k):
            list_index_na.append(j)
list_index_na = {i for i in list_index_na}
list_index_na = [i for i in list_index_na]
list_index_na.sort()
df.loc[list_index_na]
```

Now we will try to fill the NaN value for some of the columns.
For the NaN in Num-of-Doors, Bore, and  Stroke columns fill the NaN value with the mean of their columns with the similar value in the Make and Body-Style columns. *e.g if you have NaN in num-of-doors with hatchback renault car then fill the NaN with the num-of-doors mean of the hatchback renault car. 
 

```python
for col in df.columns:
    for i in list_index_na:
        if pd.isna(df.loc[i,col]):
            df.loc[i,col] = df.groupby(by=['Body-Style','Make']).mean().loc[df.loc[i,'Body-Style'],df.loc[i,'Make']][col]
list_index_na.sort()
df.loc[list_index_na]
```

Let's see once more our previous dataframe which contain the NaN to make sure that it have been filled properly

Now, drop all the rest row that contain the NaN value

```python
df.dropna(inplace=True)
df.info()
```

## Question 1

Do you think there is a relation between the price column and the Make column? Make you reasoning

```python
for i in df['Make'].unique():
    from scipy.stats import anderson
    data = df[df['Make']==i]['Price']
    result = anderson(data)
    print(i, 'x Price')
    print('stat=%.3f' % (result.statistic))
    for i in range(len(result.critical_values)):
        sl, cv = result.significance_level[i], result.critical_values[i]
        if result.statistic < cv:
            print('Probably Gaussian at the %.1f%% level' % (sl))
        else:
            print('Probably not Gaussian at the %.1f%% level' % (sl))
    print('='*10)
```

**After normality check, some of datas are not normal at all. So we conclude that all datas are not normal**

**So, we will use the Kruskal-Wallis H-test (Our significance level is 5%), where:**
- **H0: price and make have no relation.**
- **H1: price and make have relation.**

```python
from pingouin import kruskal
kruskal(data=df, dv='Price', between='Make')
```

**Because p-unc of the test < 5%, then we conclude that price and make have relation (H1)**

Now, use the IQR method to find the row outlier based on the MPG-Difference columns

```python
q1 = df['MPG-difference'].quantile(.25)
q3 = df['MPG-difference'].quantile(.75)
iqr = q3 - q1
up = q3 + 1.5 * iqr
lo = q1 - 1.5 * iqr
```

## Question 2

From the describe, and few plots that I see it seems in our outlier based on the MPG-Differences that the car is quite big. Do you agree with my assumption? give me your reasoning

```python
data = df[(df['MPG-difference'] < lo )| (df['MPG-difference'] > up )]['MPG-difference']
result = anderson(data)
print('outlier')
print('stat=%.3f' % (result.statistic))
for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < cv:
        print('Probably Gaussian at the %.1f%% level' % (sl))
    else:
        print('Probably not Gaussian at the %.1f%% level' % (sl))
data = df[(df['MPG-difference'] >= lo )| (df['MPG-difference'] <= up )]['MPG-difference']
result = anderson(data)
print('not outlier')
print('stat=%.3f' % (result.statistic))
for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < cv:
        print('Probably Gaussian at the %.1f%% level' % (sl))
    else:
        print('Probably not Gaussian at the %.1f%% level' % (sl))
```

**After normality check, the datas are not normal at all.**

**So, we will use the Kruskal-Wallis H-test (Our significance level is 5%), where:**
- **H0: outlier based on the MPG-Differences that the car is not big.**
- **H1: outlier based on the MPG-Differences that the car is big.**

```python
from scipy.stats import mannwhitneyu
data1 = df[(df['MPG-difference'] < lo )| (df['MPG-difference'] > up )]['MPG-difference']
data2 = df[(df['MPG-difference'] >= lo )| (df['MPG-difference'] <= up )]['MPG-difference']
stat, p = mannwhitneyu(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

**Because p-unc of the test < 5%, then we conclude that outlier based on the MPG-Differences that the car is big (H1)**
